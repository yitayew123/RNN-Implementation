```python
# ----------------- Import Necessary Libraries -----------------
import pandas as pd                     # For handling and processing tabular data
import matplotlib.pyplot as plt         # For creating static visualizations
import seaborn as sns                   # For advanced and beautiful statistical plots

# ----------------- Create Dataset -----------------
data = {
    'Role': ['Technical Officer', 'Issues Resolver', 'Basemap Officer', 'Document Verifier', 'Director'],
    'Count': [48, 11, 13, 18, 9]
}

# Convert dictionary to a pandas DataFrame
df = pd.DataFrame(data)

# ----------------- Add Total Row -----------------
# Calculate the total count of staff and append it to the dataset for completeness
total_value = df['Count'].sum()
df = pd.concat([df, pd.DataFrame({'Role': ['Total'], 'Count': [total_value]})], ignore_index=True)

# ----------------- Define Custom Colors -----------------
# Use soft pastel colors for visual appeal, adding an extra color for 'Total'
custom_colors = ['#FFB6C1', '#87CEEB', '#FFD700', '#98FB98', '#FF7F50', '#CBAACB']

# ----------------- Set Global Plot Style -----------------
sns.set_style("whitegrid")  # Adds clean white grid background for better readability

# ----------------- Create Figure and Subplots -----------------
fig, axes = plt.subplots(1, 2, figsize=(15, 7))  # 1 row, 2 columns for side-by-side plots

# ----------------- Add Common Title -----------------
# suptitle() is used for an overall title for all subplots
fig.suptitle("Distribution of Staff Roles and Their Proportion (Farmer Service)", 
             fontsize=18, fontweight='bold', color='darkblue', y=1.02)

# ============================================================
#                     BAR CHART (Left)
# ============================================================
barplot = sns.barplot(
    x='Role', 
    y='Count', 
    data=df, 
    palette=custom_colors, 
    ax=axes[0]
)

# Add data labels on top of each bar
for index, row in df.iterrows():
    barplot.text(
        index, 
        row['Count'] + 1,         # Position slightly above the bar
        row['Count'],             # Label text
        color='black', 
        ha='center', 
        fontsize=12, 
        fontweight='bold'
    )

# Customize bar chart appearance
axes[0].set_title("Number of Trainees Across Each Department\n(Farmer Service)", 
                  fontsize=16, fontweight='bold', pad=20, color='darkslateblue')
axes[0].set_xlabel("Role", fontsize=13, color='dimgray', labelpad=15, fontweight='bold')
axes[0].set_ylabel("Number of Staff", fontsize=13, color='dimgray', fontweight='bold')
axes[0].tick_params(axis='x', rotation=20, labelsize=11)
axes[0].grid(axis='y', linestyle='--', alpha=0.6)

# Highlight the "Total" bar for emphasis
barplot.patches[-1].set_edgecolor('black')
barplot.patches[-1].set_linewidth(2.5)

# ============================================================
#                     PIE CHART (Right)
# ============================================================
# Exclude "Total" row from pie chart for clearer visualization
df_pie = df[df['Role'] != 'Total']

# Create an "explode" effect to slightly pull slices apart
explode_values = [0.05] * len(df_pie)

# Plot pie chart
wedges, texts, autotexts = axes[1].pie(
    df_pie['Count'],
    labels=df_pie['Role'],
    autopct='%1.1f%%',          # Show percentage values
    startangle=140,             # Rotate starting angle for balance
    explode=explode_values,     # Separate slices
    shadow=True,                # Add shadow for depth
    colors=custom_colors[:-1],  # Exclude the color for 'Total'
    textprops={'fontsize': 11, 'fontweight': 'bold', 'color': 'black'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)

# Add a circle at the center to create a donut-style pie chart
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
axes[1].add_artist(centre_circle)

# Display total count in the center
axes[1].text(0, 0, f"Total\n{total_value}", fontsize=14, fontweight='bold', 
             color='darkblue', ha='center', va='center')

# Add title to pie chart
axes[1].set_title("Proportion of Trainees by Role", 
                  fontsize=16, fontweight='bold', color='darkslateblue', pad=20)

# Customize percentage text color and font
for autotext in autotexts:
    autotext.set_color('darkred')
    autotext.set_fontsize(11)
    autotext.set_fontweight('bold')

# ----------------- Final Layout Adjustments -----------------
plt.tight_layout()   # Prevent overlapping of subplots
sns.despine()        # Remove top and right borders for a clean look

# ----------------- Show Final Plots -----------------
plt.show()
#hadle warnings
import warnings
warnings.filterwarnings("ignore")
```

```python
# ----------------- Import Libraries -----------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------- Create Dataset -----------------
data = {
    'Round': ['Round 1', 'Round 2', 'Round 3', 'Round 4', 'Round 5', 'Round 6', 'Round 7', 'Round 8'],
    'Received': [9, 50, 340, 614, 126, 250, 343, 2392],
    'Migrated': [9, 37, 304, 580, 115, 227, 301, 2292],
    'Issues': [0, 13, 36, 34, 11, 23, 42, 100]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# ----------------- Calculate Totals -----------------
totals = pd.DataFrame({
    'Round': ['Total'],
    'Received': [df['Received'].sum()],
    'Migrated': [df['Migrated'].sum()],
    'Issues': [df['Issues'].sum()]
})

# Append total row to the original DataFrame
df = pd.concat([df, totals], ignore_index=True)

# ----------------- Set Style -----------------
sns.set_style("whitegrid")
plt.figure(figsize=(16, 7))

# ----------------- Plot Grouped Bar Chart -----------------
bar_width = 0.25
index = range(len(df))

# Plot bars for each category
plt.bar([i - bar_width for i in index], df['Received'], width=bar_width, label='Received', color='#87CEEB')
plt.bar(index, df['Migrated'], width=bar_width, label='Migrated', color='#32CD32')
plt.bar([i + bar_width for i in index], df['Issues'], width=bar_width, label='Issues', color='#FF6347')

# ----------------- Add Data Labels -----------------
for i in index:
    plt.text(i - bar_width, df['Received'][i] + 20, df['Received'][i], ha='center', va='bottom', fontsize=10, fontweight='bold')
    plt.text(i, df['Migrated'][i] + 20, df['Migrated'][i], ha='center', va='bottom', fontsize=10, fontweight='bold')
    plt.text(i + bar_width, df['Issues'][i] + 20, df['Issues'][i], ha='center', va='bottom', fontsize=10, fontweight='bold')

# ----------------- Customize Plot -----------------
plt.title("Data Migration Report (Round-wise) with Totals", fontsize=18, fontweight='bold', color='darkblue', pad=20)
plt.xlabel("Round", fontsize=13, fontweight='bold', color='dimgray')
plt.ylabel("Number of Records", fontsize=13, fontweight='bold', color='dimgray')
plt.xticks(index, df['Round'], rotation=30, ha='right', fontsize=11)
plt.yticks(fontsize=11)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.legend(title="Status", fontsize=11)

# ----------------- Show Plot -----------------
plt.tight_layout()
plt.show()
```

```python
# ----------------- Import Libraries -----------------
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# ----------------- Create Dataset -----------------
data = {
    'Round': ['Round 1', 'Round 2', 'Round 3', 'Round 4', 'Round 5', 'Round 6', 'Round 7', 'Round 8'],
    'Received': [9, 50, 340, 614, 126, 250, 343, 2392],
    'Migrated': [9, 37, 304, 580, 115, 227, 301, 2292],
    'Issues': [0, 13, 36, 34, 11, 23, 42, 100]
}

# Convert to DataFrame
df = pd.DataFrame(data)

# ----------------- Calculate Totals -----------------
totals = pd.DataFrame({
    'Round': ['Total'],
    'Received': [df['Received'].sum()],
    'Migrated': [df['Migrated'].sum()],
    'Issues': [df['Issues'].sum()]
})

# Append total row
df = pd.concat([df, totals], ignore_index=True)

# ----------------- Colors -----------------
received_colors = sns.color_palette("Blues_r", len(df))
migrated_colors = sns.color_palette("Greens_r", len(df))
issues_colors = sns.color_palette("Reds_r", len(df))

# ----------------- Create Subplots -----------------
fig, axes = plt.subplots(1, 2, figsize=(18, 7))
bar_width = 0.25
index = range(len(df))

# ----------------- Grouped Bar Chart -----------------
for i in index:
    axes[0].bar(i - bar_width, df['Received'][i], width=bar_width, color=received_colors[i], edgecolor='black', linewidth=1.5)
    axes[0].bar(i, df['Migrated'][i], width=bar_width, color=migrated_colors[i], edgecolor='black', linewidth=1.5)
    axes[0].bar(i + bar_width, df['Issues'][i], width=bar_width, color=issues_colors[i], edgecolor='black', linewidth=1.5)

# ----------------- Add Labels -----------------
for i in index:
    axes[0].text(i - bar_width, df['Received'][i] + 20, df['Received'][i], ha='center', va='bottom', fontsize=10, fontweight='bold', color='navy')
    axes[0].text(i, df['Migrated'][i] + 20, df['Migrated'][i], ha='center', va='bottom', fontsize=10, fontweight='bold', color='darkgreen')
    axes[0].text(i + bar_width, df['Issues'][i] + 20, df['Issues'][i], ha='center', va='bottom', fontsize=10, fontweight='bold', color='darkred')

# Customize axes
axes[0].set_title("Data Migration Report (Round-wise) with Totals", fontsize=16, fontweight='bold', color='darkblue', pad=15)
axes[0].set_xlabel("Round", fontsize=12, fontweight='bold', color='dimgray')
axes[0].set_ylabel("Number of Records", fontsize=12, fontweight='bold', color='dimgray')
axes[0].set_xticks(index)
axes[0].set_xticklabels(df['Round'], rotation=30, ha='right', fontsize=11)
axes[0].grid(axis='y', linestyle='--', alpha=0.5)
axes[0].legend(['Received', 'Migrated', 'Issues'], title="Status", fontsize=11)

# ----------------- PIE CHART -----------------
# Pie chart of totals
total_received = df['Received'].sum()
total_migrated = df['Migrated'].sum()
total_issues = df['Issues'].sum()
totals_values = [total_received, total_migrated, total_issues]
totals_labels = ['Received', 'Migrated', 'Issues']
pie_colors = ['#87CEEB', '#32CD32', '#FF6347']
explode = (0.05, 0.05, 0.05)

wedges, texts, autotexts = axes[1].pie(
    totals_values,
    labels=totals_labels,
    autopct='%1.1f%%',
    startangle=140,
    colors=pie_colors,
    explode=explode,
    shadow=True,
    textprops={'fontsize': 12, 'fontweight': 'bold', 'color': 'black'},
    wedgeprops={'edgecolor': 'white', 'linewidth': 2}
)
axes[1].set_title("Overall Distribution of Migration Status", fontsize=16, fontweight='bold', color='darkblue', pad=15)

# Donut effect
centre_circle = plt.Circle((0, 0), 0.60, fc='white')
axes[1].add_artist(centre_circle)

# ----------------- Layout -----------------
plt.suptitle("E-Carta Service: Data Migration Report", fontsize=20, fontweight='bold', color='darkgreen', y=1.05)
plt.tight_layout()
sns.despine()
plt.show()
```
